```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
#ptin(cwd)
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
working_file = os.path.join(CACHE_DIR, 'movies-ranking-dataset.csv')
ouput_file = os.path.join(CACHE_DIR, 'movies-ranking-dataset-cleaned.csv')
```

```python
df = pd.read_csv(working_file)
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace =True)
df['Rank'] = -1
df.head(10)
```

```python
df['Domestic %'] = df['%']
df['Foreign %'] = df['%.1']
```

```python
df.head()
```

```python
df = df.drop(['%', '%.1'], axis = 1)
```

```python
df.head()
```

```python
df['Worldwide'] = df['Worldwide'].replace('$','').replace(',','')
df.head()
```

## Clean Col

```python
to_clean_cols = ['Worldwide', 'Domestic', 'Foreign']
def currency_str_to_int(current_val):
    currency_val = current_val.replace('$','').replace(',','')
    try:
        currency_val = int(currency_val)
    except:
        currency_val = 0
    return currency_val

def clean_cols(row):
    for col in to_clean_cols:
        current_val = row[col]
        row[col] = currency_str_to_int(current_val)
    return row
```

```python
df_cleaned = df.apply(clean_cols, axis=1)
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.sort_values(by=['Worldwide'], inplace =True, ascending=False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace =True, drop= True)
df_cleaned.head()
```

```python
df_cleaned['Rank'] = df_cleaned.index+1
df_cleaned
```

```python
df_cleaned['Domestic %'] = df_cleaned['Domestic'] /df_cleaned['Worldwide']
df_cleaned['Foreign %'] = df_cleaned['Foreign'] / df_cleaned['Worldwide']
```

```python
df_cleaned.head()
```

```python
df_cleaned.to_csv(ouput_file, index = False)
```
