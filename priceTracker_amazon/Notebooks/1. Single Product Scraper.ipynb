```python
import requests
from requests_html import HTML
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.add_argument('--headless')

driver = webdriver.Chrome(options = options)
```

```python
url = 'https://www.amazon.com/WAVLINK-Thunderbolt-Delivery-DisplayPort-Ethernet/dp/B07N6552YK/ref=sr_1_1_sspa?dchild=1&keywords=thunderbolt%2Bdocking%2Bstation&qid=1589158487&sr=8-1-spons&spLa=ZW5jcnlwdGVkUXVhbGlmaWVyPUE2RjhNQjE1MkFKS0omZW5jcnlwdGVkSWQ9QTAxMTQxMjYxNlVHWkFNNEw2TjRQJmVuY3J5cHRlZEFkSWQ9QTA0ODQ4MDMzRDRESU5SWTdVMTlEJndpZGdldE5hbWU9c3BfYXRmJmFjdGlvbj1jbGlja1JlZGlyZWN0JmRvTm90TG9nQ2xpY2s9dHJ1ZQ&th=1'
title_lookup = '#productTitle'
price_lookup = '#priceblock_ourprice'
```

```python
driver.get(url)
body_el = driver.find_element_by_css_selector('body')
html_str = body_el.get_attribute('innerHTML')
print(html_str)
```

```python
html_obj = HTML(html=html_str)
```

```python
product_title = html_obj.find(title_lookup, first = True).text
```

```python
product_price = html_obj.find(price_lookup, first=True).text
```

```python
print(product_title, product_price)
```
